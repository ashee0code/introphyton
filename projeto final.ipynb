```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```



###### PERGUNTA 2 

###### QUAL A RELAÇÃO ENTRE A POPULARIDADE DAS MUSICAS E AS SUAS DURAÇÕES?

```python
df=pd.read_csv("Spotify2000-2023.csv", sep=";")


df.info()
df["duration"]=pd.to_numeric(df["duration"],errors="coerce")

df["popularity"]=pd.to_numeric(df["popularity"],errors="coerce")

df["artist"]=df["artist"].str.lower().str.strip()
```

### PERGUNTA 1

#### QUAL A RELAÇÃO ENTRE A POPULARIDADE DAS MUSICAS E AS SUAS DURAÇÕES?

```python
#Gráfico dataset 
plt.bar((df["duration"]),df["popularity"])
plt.title("Gráfico musical")
plt.xlabel("Duração")
plt.ylabel("Popularidade")
plt.show()
```

### RESPOSTA:

### No gráfico, a duração tem relação sim com a popularidade visto que a maioria das musicas populares tem uma média parecida de duração. Além disso, músicas com menores durações tendem a fazer mais sucesso entre a geração Z baseado no aumento de dopamina gerada pela descoberta de musicas novas.
